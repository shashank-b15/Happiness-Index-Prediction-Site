## Pre-processing

```python
from sklearn.preprocessing import LabelEncoder 

lab=LabelEncoder()

data['Quality']=lab.fit_transform(data['Quality'])
data['Teacher_Quality']=lab.fit_transform(data['Teacher_Quality'])
data['Books']=lab.fit_transform(data['Books'])
data['Classroom']=lab.fit_transform(data['Classroom'])
data['Food']=lab.fit_transform(data['Food'])
data['Homework']=lab.fit_transform(data['Homework'])
data['Continue_school']=lab.fit_transform(data['Continue_school'])
data['Understanding_subs']=lab.fit_transform(data['Understanding_subs'])
data['Meetings']=lab.fit_transform(data['Meetings'])
```

## Data Splitting

```python
x=data.iloc[:,data.columns!='Happiness']
y=data.iloc[:,data.columns=='Happiness']

#x.shape

#x.head()

x.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix,recall_score, roc_curve, auc
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3)

xtrain.head()

ytrain.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import AdaBoostRegressor,GradientBoostingRegressor,RandomForestRegressor
from sklearn.ensemble import StackingRegressor
#base models
base_models=[
    ('ABR',AdaBoostRegressor(n_estimators=50)),
    ('GBR',GradientBoostingRegressor(n_estimators=50)),
    ('RFR',RandomForestRegressor(n_estimators=50))
]
#Final Model
final_model=LinearRegression()

#stacking Regressor
classifier=StackingRegressor(estimators=base_models,final_estimator=final_model)
classifier.fit(xtrain,ytrain)
ypred=classifier.predict(xtest)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
#Evaluation of Model
print("R2 Score " ,r2_score(ytest,ypred))
mse=mean_squared_error(ytest,ypred)
mae=mean_absolute_error(ytest,ypred)
print("MSE:",mse)
print("MAE:",mae)
print("RMSE :",np.sqrt(mse))
```

```python
import pickle
filename = 'parent.sav'
pickle.dump(classifier, open(filename, 'wb'))
# loading the saved model
loaded_model = pickle.load(open('parent.sav', 'rb'))
```
